# 1.4 텐서(Tensor)
* 다수의 행렬이 모이면 텐서
* 배열이나 행렬과 매우 유사한 특수 자료구조
* 파이토치는 텐서를 사용하여 모델의 입력과 출력, 모델의 매개변수들을 처리

<center><img src='https://miro.medium.com/max/875/1*jRyyMAhS_NZxqyv3EoLCvg.png' width=500></center>

```python
tensor1 = torch.tensor([[[1, 2], [3, 4]], [[5, 6], [7, 8]]])
print(tensor1)
```

```python
tensor2 = torch.tensor([[[9, 10], [11, 12]], [[13, 14], [15, 16]]])
print(tensor2)
```

```python
# 두 텐서의 사칙 연산
print(tensor1 + tensor2)
print(tensor1 - tensor2)
print(tensor1 * tensor2)
print(tensor1 / tensor2)
```

```python
print(torch.add(tensor1, tensor2))      # 더하기
print(torch.subtract(tensor1, tensor2)) # 빼기
print(torch.multiply(tensor1, tensor2)) # 곱하기
print(torch.divide(tensor1, tensor2))   # 나누기
print(torch.matmul(tensor1, tensor2))   # 행렬곱
```

```python
# tensor1에 결과를 다시 저장
# 모든 사칙 연산자에 _를 붙이면 inplace=True가 됨
print(tensor1.add_(tensor2))
print(tensor1)
```

# 2. 텐서의 변환

```python
data = [[1, 2], [3, 4]]
print(data)
```

```python
x_data = torch.tensor(data)
print(x_data)
```

```python
import numpy as np
```

```python
np_array = np.array(x_data)
np_array
```

```python
x_np_1 = torch.tensor(np_array)
x_np_1
```

```python
x_np_1[0, 0] = 100
print(x_np_1)
print(np_array)
```

```python
# as_tensor: ndarray와 동일한 메모리 주소를 가리키는 뷰를 만듦
x_np_2 = torch.as_tensor(np_array)
print(x_np_2)
x_np_2[0, 0] = 200
print(x_np_2)
print(np_array)
```

```python
# from_numpy(): ndarray와 동일한 메모리 주소를 가리키는 뷰를 만듦
x_np_3 = torch.from_numpy(np_array)
print(x_np_3)
x_np_3[0, 1] = 400
print(x_np_3)
print(np_array)
```

```python
np_again = x_np_1.numpy()
print(np_again, type(np_again))
```

# 5. 코랩에서 GPU 사용하기
* 코랩에서 device를 변경하는 방법
  * 상단 메뉴 -> 런타임 -> 런타임 유형 변경 -> 하드웨어 가속기를 GPU로 변경 -> 저장 -> 세션 다시 시작 및 모두 실행

```python
tensor = torch.rand(3, 4)
print(f'shape: {tensor.shape}')
print(f'dtype: {tensor.dtype}')
print(f'device: {tensor.device}')
```

```python
# is_available(): gpu 사용할 수 있는 여부
tensor = tensor.reshape(4, 3)
tensor = tensor.int()
if torch.cuda.is_available():
  print('GPU를 사용할 수 있음')
  tensor = tensor.to('cuda')

print(f'shape: {tensor.shape}')
print(f'dtype: {tensor.dtype}')
print(f'device: {tensor.device}')
```
