# 1. Iris DataSet
* 데이터셋: 특정한 작업을 위해 데이터를 관련성 있게 모아놓은 것
* [사이킷런 데이터셋 페이지](https://scikit-learn.org/stable/modules/classes.html?highlight=datasets#module-sklearn.datasets)

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
iris
```

```python
print(iris['DESCR'])
```

* sepal length in cm: 꽃받침의 길이
* sepal width in cm: 꽃받침의 너비
* petal length in cm: 꽃잎의 길이
* petal width in cm: 꽃잎의 너비

```python
data = iris['data']
data
```

```python
feature_names = iris['feature_names']
feature_names
```

```python
import pandas as pd
```

```python
df_iris = pd.DataFrame(data, columns=feature_names)
df_iris
```

```python
target = iris['target']
target
```

```python
df_iris['target'] = target
df_iris
```

```python
from sklearn.model_selection import train_test_split
```

```python
# train_test_split(독립변수, 종속변수, 테스트사이즈, 시드값 ...)
X_train, X_test, y_train, y_test = train_test_split(df_iris.drop('target', 1), df_iris['target'], test_size=0.2, random_state=2023)
```

```python
X_train.shape, X_test.shape
```

```python
y_train.shape, y_test.shape
```

```python
X_train
```

```python
y_train
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```python
svc = SVC()
```

```python
svc.fit(X_train, y_train)
```

```python
y_pred = svc.predict(X_test)
```

```python
y_pred
```

```python
print('정답률', accuracy_score(y_test, y_pred))
```

```python
# test 데이터 생성: 6.2, 2.2, 4.3, 1.4
y_pred = svc.predict([[6.2, 2.2, 4.3, 1.4]])
y_pred
```
