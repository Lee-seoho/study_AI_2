```python
import matplotlib.pyplot as plt
import numpy as np

# 첫 번째 데이터 세트
data1 = np.arange(1, 50)
data3 = np.arange(50, 100)

# 두 번째 데이터 세트
data2 = np.arange(50, 100)

# 첫 번째 서브플롯 생성 및 두 개의 선 그리기
plt.subplot(2, 1, 1)
plt.plot(data1, label='Line 1')
plt.plot(data3, label='Line 2')
plt.legend()

# 두 번째 서브플롯 생성 및 한 개의 선 그리기
plt.subplot(2, 1, 2)
plt.plot(data2, label='Line 3')
plt.legend()

plt.show()
```

```python
plt.rc('font', family='NanumBarunGothic')
```

```python
plt.figure(figsize=(6, 8)) # inch
plt.plot([1, 2, 3], [1, 2, 3])
plt.plot([1, 2, 3], [2, 4, 6])
plt.title('제목', fontsize=30)
plt.xlabel('x축', fontsize=20)
plt.ylabel('y축', fontsize=20, rotation=30)
```

```python
plt.figure(figsize=(15, 10))
plt.title('마커설정', fontsize=30)
plt.plot(np.arange(10), np.arange(10), color='deepskyblue', marker='o', ms='10',linestyle='-.')
plt.plot(np.arange(10), np.arange(10)*2, color='deeppink', marker='v',  linestyle='--')
plt.plot(np.arange(10), np.arange(10)*3, color='gold', marker='*', ms='30', linestyle='')

# 범례: 기본 위치는 왼쪽 상단
plt.legend(['10', '10 * 2', '10 * 3'], fontsize=15, loc='lower right', ncol=3)

plt.xlim(0, 12)
plt.ylim(0, 30)

plt.xticks(rotation=30)
plt.yticks(rotation=30)

plt.grid()
plt.show()
```

```python
x = ['파이썬', '데이터분석', '머신러닝', '딥러닝', '자연어처리', '컴퓨터비전']
y = [95, 80, 75, 50, 40, 25]
plt.figure(figsize=(8,5))
plt.title('AI 성적표', fontsize=25)
plt.ylabel('수강생 점수')
plt.bar(x, y, alpha=0.4, color='deeppink')
plt.show()
```

```python
x = ['파이썬', '데이터분석', '머신러닝', '딥러닝', '자연어처리', '컴퓨터비전']
y = [95, 80, 75, 50, 40, 25]
plt.figure(figsize=(8,5))
plt.title('AI 성적표', fontsize=25)
plt.ylabel('수강생 점수')
plt.barh(x, y, alpha=0.4, color='deepskyblue')
plt.show()
```
